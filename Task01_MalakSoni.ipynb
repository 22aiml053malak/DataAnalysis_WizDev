```python
# Aggregate data by UPI APP
aggregated_data = data.groupby("UPI APP").agg({
    "VOL(Million)": "sum"  # Total transaction volume
}).reset_index()

# Create a new DataFrame with aggregated data
aggregated_df = pd.DataFrame(aggregated_data)
#We groupped our UPI APPS For creating a category
```

```python

# Analyze the aggregated data
# transaction volumes for each UPI APP category
plt.figure(figsize=(180, 80))
plt.bar(aggregated_df['UPI APP'], aggregated_df['VOL(Million)'])
plt.xticks(rotation=90)
plt.xlabel('UPI App')
plt.ylabel('Total Transaction Volume (Million)')
plt.title('Total Transaction Volume by UPI App')
plt.show()
```

```python
# Remove commas from the 'VOL(Million)' column in your 'data' DataFrame beacuse there can be error
data['VOL(Million)'] = data['VOL(Million)'].str.replace(',', '')

# Convert the 'VOL(Million)' column to float
data['VOL(Million)'] = data['VOL(Million)'].astype(float)

# Aggregate data by UPI APP
aggregated_data = data.groupby("UPI APP").agg({
    "VOL(Million)": "sum" 
}).reset_index()

# Sort aggregated data by transaction volumes in descending order
aggregated_df_sorted = aggregated_data.sort_values(by='VOL(Million)', ascending=False)

```

```python
# Plot the Bar Chart of Total Transaction ****Volume v/s UPI App*****
plt.figure(figsize=(48, 36))
plt.bar(aggregated_df_sorted['UPI APP'], aggregated_df_sorted['VOL(Million)'])
plt.xticks(rotation=45)
plt.xlabel('UPI App')
plt.ylabel('Total Transaction Volume (Million)')
plt.title('Total Transaction Volume by UPI App')
plt.tight_layout()
plt.show()
```

```python
data['VOL(Million)'] = data['VOL(Million)'].astype(float)

# Grouping based on 'UPI APP' and 'DATE'
# Assuming every 50 or more rows have the same 'DATE'
# You may need to adjust the group size based on your actual data
group_size = 50
data['Group'] = (data.index // group_size) + 1

# Aggregate data by 'UPI APP', 'DATE', and 'Group'
aggregated_data = data.groupby(["UPI APP", "DATE", "Group"]).agg({
    "VOL(Million)": "sum"  # Total transaction volume
}).reset_index()

# Sort aggregated data by 'DATE'
aggregated_data_sorted = aggregated_data.sort_values(by='DATE')

# Create a list of unique UPI Apps
unique_apps = aggregated_data_sorted['UPI APP'].unique()

# Plot market capture for each UPI App over time
plt.figure(figsize=(12, 8))
for app in unique_apps:
    app_data = aggregated_data_sorted[aggregated_data_sorted['UPI APP'] == app]
    plt.plot(app_data['DATE'], app_data['VOL(Million)'], label=app)

plt.xlabel('Date')
plt.ylabel('Total Transaction Volume (Million)')
plt.title('Market Capture of UPI Apps over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
colors = {'PhonePe': 'blue', 'Google Pay': 'green', 'Paytm': 'red'}

# Plot market capture for each UPI App over time
#For conviniance i highlighted three of our highest category in diff col...
plt.figure(figsize=(12, 8))
for app in unique_apps:
    app_data = aggregated_data_sorted[aggregated_data_sorted['UPI APP'] == app]
    if app in colors:
        plt.plot(app_data['DATE'], app_data['VOL(Million)'], label=app, color=colors[app])
    else:
        # Use default color for other apps
        plt.plot(app_data['DATE'], app_data['VOL(Million)'], label=app)

plt.xlabel('Date')
plt.ylabel('Total Transaction Volume (Million)')
plt.title('Market Capture of UPI Apps over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
phonepe_data = data[data['UPI APP'] == 'PhonePe']['VOL(Million)'].sum()
googlepay_data = data[data['UPI APP'] == 'Google Pay']['VOL(Million)'].sum()
paytm_data = data[data['UPI APP'] == 'Paytm']['VOL(Million)'].sum()

# Calculate total transaction volume for all three apps
total_volume = phonepe_data + googlepay_data + paytm_data

# Calculate market share for each app
phonepe_share = (phonepe_data / total_volume) * 100
googlepay_share = (googlepay_data / total_volume) * 100
paytm_share = (paytm_data / total_volume) * 100

# Create labels and sizes for the pie chart
labels = ['PhonePe', 'Google Pay', 'Paytm']
sizes = [phonepe_share, googlepay_share, paytm_share]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Market Share of PhonePe, Google Pay, and Paytm,Created By "Malak Soni"')
plt.axis('equal')  
plt.show()
#This pie chart is to Analize  market share between three of them we can also check for all but i think this is enough
```
