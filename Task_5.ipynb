```python
# plot the each category total amount of 2018
data_2018=df[df['Year']==2018]
grouped_df = data_2018.groupby('Payment_Mode')['Total_Amount'].sum()
# Plot pie chart for each payment mode
grouped_df.plot(kind='pie', autopct='%1.1f%%', startangle=140, figsize=(10, 6))
plt.title('Distribution of Payment Modes in 2018')
plt.ylabel('')
plt.show()
```
